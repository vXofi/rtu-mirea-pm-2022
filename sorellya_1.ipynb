4.3.2 Задание

Дана функция sqrt(1 + e**sqrt(x) + cos(x**2))/abs(1 - sin(x)**3) + log(abs(2x))
Создать массив из 10 значений функции. Выделить срез первой половины массива и построить графики для основного массива - линейный и для среза - точечный

```python
def func(x):
  return np.sqrt(1 + np.e**np.sqrt(x) + np.cos(x**2))/abs(1 - np.sin(x)**3) + np.log(abs(2*x))

x_values = range(1, 11)
x_zrez = x_values[:len(x_values)//2]
y_values = [func(i) for i in x_values]

figure, axs = plt.subplots(2, figsize = (4, 4))
axs[0].plot(x_values, y_values)
axs[1].scatter(x_zrez, y_values[:len(x_zrez)], color="red")
```

4.3.3 Задание

Дана функция |cos(𝑥𝑒**
(cos(𝑥)+ln(𝑥+1))
)| Построить график на
интервале (0,10) с шагом 1 с заливкой площади и найти эту площадь
под ним

```python
from scipy.integrate import simps

def func(x):
  return np.abs(np.cos(x * np.e**(np.cos(x) + np.log(x + 1))))

x = np.arange(0, 10, 1)
y = func(x)
plt.grid()
plt.plot(x, y)
plt.fill_between(x, y)
area = np.trapz(y)
print(area)
```
