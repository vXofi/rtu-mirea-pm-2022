3.3.2 Задание

Задача:
Загрузить фрейм данных по ссылке:
https://raw.githubusercontent.com/akmand/datasets/master/iris.csv.
Необходимо выполнить нормализацию первого числового признака
(sepal_length_cm) с использованием минимаксного преобразования, а
второго (sepal_width_cm) с задействованием z-масштабирования.


```python
from sklearn import preprocessing
dataset = pd.read_csv("https://raw.githubusercontent.com/akmand/datasets/master/iris.csv", sep=",")
```

```python
scaler = preprocessing.MinMaxScaler()
dataset["sepal_length_cm"] = scaler.fit_transform(dataset["sepal_length_cm"].values.reshape(-1, 1))
mid = np.mean(dataset["sepal_width_cm"])
st_div = np.std(dataset["sepal_width_cm"])
dataset["sepal_width_cm"] = (dataset["sepal_width_cm"] - mid) / st_div
dataset.head()
```
